# I. Data Preparation
---

```python
# Initial imports

import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path

%matplotlib inline

# Set the path to CSV file
whale_returns_path = Path("/Users/coneychiwabee/fintech_uoftbootcamp/pandas-homework/Starter_Code/Resources/whale_returns.csv")
algo_returns_path = Path("/Users/coneychiwabee/fintech_uoftbootcamp/pandas-homework/Starter_Code/Resources/algo_returns.csv")
sp_tsx_history_path = Path("/Users/coneychiwabee/fintech_uoftbootcamp/pandas-homework/Starter_Code/Resources/sp_tsx_history.csv")

# Read in the CSVs as a DataFrames
whale_returns = pd.read_csv(whale_returns_path, index_col="Date",parse_dates=True, infer_datetime_format=True)
algo_returns = pd.read_csv(algo_returns_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
sp_tsx_history = pd.read_csv(sp_tsx_history_path,index_col="Date", parse_dates=True, infer_datetime_format=True)


#whale_returns.head()
#algo_returns.head()
sp_tsx_history.head()
```

```python
# Sort dates in ascending value and drop null values

whale_returns = whale_returns.sort_index().dropna().copy()
algo_returns = algo_returns.sort_index().dropna().copy()
sp_tsx_history = sp_tsx_history.sort_index().dropna().copy()

whale_returns.head()
#algo_returns.head()
#sp_tsx_history.head()
```

```python
# Detect null values
whale_returns.isnull().mean() * 100
whale_returns.isnull().sum()

algo_returns.isnull().mean() * 100
algo_returns.isnull().sum()

sp_tsx_history.isnull().mean() * 100
sp_tsx_history.isnull().sum()
#sp_tsx_history.head()

sp_tsx_history.dtypes
```

```python
# Remove $ 
sp_tsx_history['Close'] = sp_tsx_history['Close'].str.replace('$', '')
sp_tsx_history.head()
```

```python
#Convert close to from object to float
#Code Reference: https://stackoverflow.com/questions/39125665/cannot-convert-string-to-float-in-pandas-valueerror

sp_tsx_history = pd.concat(
                [sp_tsx_history[col].str.split()
                .str[0]
                .str.replace(',','').astype(float) for col in sp_tsx_history], axis=1)
sp_tsx_history.head()
#sp_tsx_history.dtypes
```

```python
# Convert the S&P TSX 60 closing prices to daily returns.
sp_tsx_history['S&P TSX 60 Returns'] = sp_tsx_history.pct_change(periods = 1)
sp_tsx_history.head()
```

```python
# Drop "Close" column
sp_tsx_returns = sp_tsx_history.drop(columns=['Close'])
sp_tsx_returns.head()
```

```python
# Combine all dataframes in one dataframe
combined_returns_df = pd.concat(
    [whale_returns, algo_returns, sp_tsx_returns], axis="columns", join="inner")
combined_returns_df
```

## E. Sharpe Ratios

```python
#Sharpe Ratios
# Code Reference: https://towardsdatascience.com/calculating-sharpe-ratio-with-python-755dcb346805

# Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot.
sharpe_ratios = (combined_returns_df.mean() / (combined_returns_df.std()) * np.sqrt(252))
sharpe_ratios
```

```python
sharpe_ratios.plot(kind="bar", title="Sharpe Ratios")
```

```python
snp_sharpe_ratios_avg = sharpe_ratios.mean()
print (snp_sharpe_ratios_avg)
```

### Rolling 60-Day Beta vs. S&P TSX 60

```python
# Calculate and plot the 60-day rolling beta for your portfolio compared to the S&P 60 TSX.

# Combine returns of My Portfolio and S&P 60 TSX
my_portfolio_snp60_returns = pd.concat([my_portfolio_returns, sp_tsx_returns], axis="columns", join="inner")
my_portfolio_snp60_returns.head().sort_index()
```

```python
# Calculate covariance of My Portfolio's returns vs. S&P 60 TSX
blk_covariance = my_portfolio_snp60_returns["BlackRock, Inc."].cov(my_portfolio_snp60_returns["S&P TSX 60 Returns"])
jef_covariance = my_portfolio_snp60_returns["Jefferies Financial Group Inc."].cov(my_portfolio_snp60_returns["S&P TSX 60 Returns"])
slf_covariance = my_portfolio_snp60_returns["Sun Life Financial Inc."].cov(my_portfolio_snp60_returns["S&P TSX 60 Returns"])

# Display the covariance of each stock
print(f"BlackRock covariance: {blk_covariance}")
print(f"Jefferies covariance: {jef_covariance}")
print(f"Sun Life covariance: {slf_covariance}")
```

```python
 # Calculate variance of My Portfolio's returns vs. S&P 500
variance = my_portfolio_snp60_returns["S&P TSX 60 Returns"].var()

# Diplay S&P 500 variance
print(f"S&P TSX 60 variance: {variance}")
```

```python
 # Calculate betas of My Portfolio's returns
blk_beta = blk_covariance / variance
jef_beta = jef_covariance / variance
slf_beta = slf_covariance / variance

# Display the beta value of each social media stock
print(f"BLK: {blk_beta} | JEF: {jef_beta} | SLF: {slf_beta}")
```

```python
# Calculate 60-day rolling covariance of My Portfolio's vs. S&P 60 TSX
rolling_blk_covariance = my_portfolio_snp60_returns["BlackRock, Inc."].rolling(window=60).cov(my_portfolio_snp60_returns["S&P TSX 60 Returns"])
rolling_jef_covariance = my_portfolio_snp60_returns["Jefferies Financial Group Inc."].rolling(window=60).cov(my_portfolio_snp60_returns["S&P TSX 60 Returns"])
rolling_slf_covariance = my_portfolio_snp60_returns["Sun Life Financial Inc."].rolling(window=60).cov(my_portfolio_snp60_returns["S&P TSX 60 Returns"])

# Calculate 30-day rolling variance of S&P 500
rolling_variance = my_portfolio_snp60_returns["S&P TSX 60 Returns"].rolling(window=60).var()

# Calculate 30-day rolling beta of social media stocks and plot the data
rolling_blk_beta = rolling_blk_covariance / rolling_variance
rolling_jef_beta = rolling_jef_covariance / rolling_variance
rolling_slf_beta = rolling_slf_covariance / rolling_variance

print(rolling_blk_beta,rolling_jef_beta,rolling_slf_beta)
```

```python
# Plot Rolling 60-Day Betas
ax = rolling_blk_beta.plot(figsize=(20, 10), title="Rolling 60-Day Beta of My Portfolios vs. S&P TSX 60")
rolling_jef_beta.plot(ax=ax)
rolling_slf_beta.plot(ax=ax)

# Set the legend of the figure
ax.legend(["BlackRock", "Jefferies", "Sun Life"])
```

<div class="alert alert-block alert-success">
    
<b>My Portfolio's Volatility versus S&P TSX 60<b>
    
Among the three firms in my portfolio, <b>Jefferies Financial Group Inc.<b> was the most volatile compared to the market. 
</div>
